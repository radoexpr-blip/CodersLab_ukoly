```python
%matplotlib ipympl

import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.plot(x, y, label='y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interaktivn graf')
plt.legend()
plt.show()
```
