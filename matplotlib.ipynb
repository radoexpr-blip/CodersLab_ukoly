```python
%matplotlib ipympl

import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.plot(x, y, label='y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interaktivní graf')
plt.legend()
plt.show()
```

```python
%matplotlib ipympl

import matplotlib.pyplot as plt

produkty = ['Jablka', 'Hrušky', 'Banány', 'Pomeranče']
prodeje = [25, 4, 30, 10]

plt.bar(produkty, prodeje, color='skyblue')
plt.xlabel('Produkt')
plt.ylabel('Počet prodaných kusů')
plt.title('Prodeje ovoce')
plt.show()
```
