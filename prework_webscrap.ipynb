```python
import requests
from bs4 import BeautifulSoup

# URL testovací stránky
url = "https://webscraper.io/test-sites/e-commerce/static"

# Volitelně přidáme hlavičku, aby nás server neblokoval
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# Stáhneme HTML obsah stránky
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Najdeme všechny produkty
produkty = soup.select(".thumbnail")

# Vytáhneme název, cenu a popis každého produktu
for produkt in produkty:
    nazev = produkt.select_one(".title").text.strip()
    cena = produkt.select_one(".price").text.strip()
    popis = produkt.select_one(".description").text.strip()
    print(f"Název: {nazev}\nCena: {cena}\nPopis: {popis}\n---")
```

```python
import matplotlib.pyplot as plt

x = [0,1,2,3,4,5]
y1 = [0,1,2,3,4,5]
y2 = [0,1,4,9,16,25]

plt.plot(x, y1, label='Plot 1')
plt.plot(x, y2, label='Plot 2')  
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
%matplotlib ipympl

import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.plot(x, y, label='y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interaktivní graf')
plt.legend()
plt.show()

```
